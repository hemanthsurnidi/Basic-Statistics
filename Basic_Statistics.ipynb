**THEORY**

1. What is statistics, and why is it important?

Statistics is the science of collecting, analyzing, interpreting, and presenting data. It helps in making informed decisions based on evidence.

2. What are the two main types of statistics?

Descriptive Statistics (summarizes data) and Inferential Statistics (draws conclusions from data).

3. What are descriptive statistics?

They summarize and organize data using measures like mean, median, mode, and standard deviation.

4. What is inferential statistics?

It involves using a sample to make inferences or predictions about a larger population.

5. What is sampling in statistics?

Sampling is selecting a subset of a population to estimate characteristics of the whole population.

6. What are the different types of sampling methods?

Simple random, stratified, systematic, cluster, convenience, and purposive sampling.

7. What is the difference between random and non-random sampling?

Random sampling gives each member an equal chance of selection; non-random relies on judgment or convenience.

8. Define and give examples of qualitative and quantitative data.

Qualitative: descriptive (e.g., color, brand); Quantitative: numerical (e.g., age, height).

9. What are the different types of data in statistics?

Qualitative: nominal and ordinal; Quantitative: discrete and continuous.

10. Explain nominal, ordinal, interval, and ratio levels of measurement.

Nominal: categories (e.g., gender); Ordinal: ranked order (e.g., satisfaction); Interval: equal intervals, no true zero (e.g., Celsius); Ratio: has true zero (e.g., weight).

11. What is the measure of central tendency?

A single value that represents the center or average of a datasetâ€”mean, median, or mode.

12. Define mean, median, and mode.

Mean: average; Median: middle value; Mode: most frequent value.

13. What is the significance of the measure of central tendency?

It provides a quick summary of a dataset and helps compare different datasets.

15. What is standard deviation, and why is it important?

Itâ€™s the square root of variance and shows the typical distance of values from the mean, making it easier to interpret variability.

16. Define and explain the term range in statistics.

Range is the difference between the maximum and minimum values in a dataset.

17. What is the difference between variance and standard deviation?

Variance is in squared units; standard deviation is the square root of variance and in original units.

18. What is skewness in a dataset?

Skewness measures the asymmetry of a distribution.

19. What does it mean if a dataset is positively or negatively skewed?

Positive skew: long right tail (mean > median); Negative skew: long left tail (mean < median).

20. Define and explain kurtosis.

Kurtosis measures the heaviness of tails in a distributionâ€”high kurtosis = more outliers; low = fewer.

21. What is the purpose of covariance?

Covariance shows whether two variables tend to move in the same or opposite directions.

22. What does correlation measure in statistics?

Correlation measures the strength and direction of a linear relationship between two variables.

23. What is the difference between covariance and correlation?

Covariance is unstandardized and shows direction; correlation is standardized (between -1 and 1) and shows strength and direction.

Used in business forecasting, medical research, government planning, sports analytics, finance, and education.

```python
# Question 1: How do you calculate the mean, median, and mode of a dataset?

import numpy as np
from scipy import stats

data = [12, 15, 21, 12, 19, 23, 12, 18, 21, 20]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Question 2: Write a Python program to compute the variance and standard deviation of a dataset

import numpy as np

data = [4, 8, 6, 5, 3, 7, 10]

variance = np.var(data)
std_dev = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Question 4: Implement sampling techniques like random sampling and stratified sampling

import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.DataFrame({
    "ID": range(1, 21),
    "Group": ["A"]*10 + ["B"]*10
})

# Simple Random Sampling
random_sample = df.sample(n=5)

# Stratified Sampling
stratified_sample, _ = train_test_split(df, test_size=0.75, stratify=df['Group'])

print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
# Question 5: Write a Python function to calculate the range of a dataset

def calculate_range(data):
    return max(data) - min(data)

dataset = [12, 25, 7, 9, 30, 18]
print("Range:", calculate_range(dataset))
```

```python
# Question 8: Generate a dataset and demonstrate positive and negative skewness

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Positive skew
data_pos = np.random.exponential(scale=2, size=1000)

# Negative skew (by inverting)
data_neg = -np.random.exponential(scale=2, size=1000)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data_pos, bins=30, kde=True, ax=axs[0])
axs[0].set_title("Positively Skewed")

sns.histplot(data_neg, bins=30, kde=True, ax=axs[1])
axs[1].set_title("Negatively Skewed")

plt.show()
```

```python
# Question 9: Write a Python script to calculate covariance between two datasets

import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 4, 6, 7, 8]

cov_matrix = np.cov(x, y)
print("Covariance between x and y:", cov_matrix[0][1])
```

```python
# Question 10: Write a Python script to calculate the correlation coefficient between two datasets

import numpy as np

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

correlation = np.corrcoef(x, y)
print("Correlation coefficient between x and y:", correlation[0][1])
```

```python
# Question 12: Implement and compare simple random sampling and systematic sampling

import pandas as pd
import numpy as np

data = pd.DataFrame({"ID": np.arange(1, 21)})

# Simple random sampling
random_sample = data.sample(n=5)

# Systematic sampling
step = len(data) // 5
systematic_sample = data.iloc[::step]

print("Random Sample:\n", random_sample)
print("\nSystematic Sample:\n", systematic_sample)
```

```python
# Question 13: Calculate the mean, median, and mode of grouped data

import pandas as pd

# Class intervals and frequencies
data = pd.DataFrame({
    "Class": ["0-10", "10-20", "20-30", "30-40"],
    "Frequency": [5, 8, 15, 7]
})

# Midpoints
data["Midpoint"] = [5, 15, 25, 35]
data["f*x"] = data["Frequency"] * data["Midpoint"]

mean = data["f*x"].sum() / data["Frequency"].sum()
print("Grouped Mean:", mean)
```

```python
# Question 17: Calculate the interquartile range (IQR) of a dataset

import numpy as np

data = [10, 15, 14, 18, 21, 23, 25, 30, 35, 50]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
```

```python
# Question 18: Implement Z-score normalization and explain its significance

import numpy as np

data = [10, 20, 30, 40, 50]
mean = np.mean(data)
std = np.std(data)

z_scores = [(x - mean) / std for x in data]

print("Z-score Normalization:", z_scores)
```

```python
# Question 20: Write a Python program to visualize covariance using a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 4, 5, 4, 5],
    'Z': [5, 3, 2, 4, 1]
})

cov_matrix = df.cov()

sns.heatmap(cov_matrix, annot=True, cmap="YlGnBu")
plt.title("Covariance Heatmap")
plt.show()
```

```python
# Question 21: Use seaborn to create a correlation matrix for a dataset

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Math': [78, 85, 96, 80, 72],
    'Science': [88, 90, 94, 82, 70],
    'English': [70, 65, 80, 75, 60]
})

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Question 24: Implement the Pearson and Spearman correlation coefficients for a dataset

from scipy.stats import pearsonr, spearmanr

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 60]

pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
